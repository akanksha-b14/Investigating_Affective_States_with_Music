```python
import pandas as pd
import numpy as np
```

```python
music_set = pd.read_csv('Music Sets - Chinese.csv')
linked_songs = pd.read_csv('linked SONGS - CHINESE.csv')
```

```python
sad_df = music_set[["SET","Sad"]]
happy_df = music_set[["SET","Happy"]]
energetic_df = music_set[["SET","Energetic"]]
```

```python
sad_df = sad_df.rename(columns={"Sad":"Code"})
happy_df = happy_df.rename(columns={"Happy":"Code"})
energetic_df = energetic_df.rename(columns={"Energetic":"Code"})
```

```python
song_features_sad = pd.merge(sad_df,linked_songs,left_on="Code",right_on="Code")
song_features_happy = pd.merge(happy_df,linked_songs,left_on="Code",right_on="Code")
song_features_energetic = pd.merge(energetic_df,linked_songs,left_on="Code",right_on="Code")
```

```python
song_features_energetic
```

```python
#Load participants list with tagged sets
participants = pd.read_csv('Participants_List - Chinese.csv')
```

```python
participants = participants.drop(['LINK','LEAD','Unnamed: 4'],axis=1)
```

```python
participants = participants[~participants.isnull().any(axis=1)]
```

```python
participant_responses = pd.read_csv('final_survey_responses.csv')
```

```python
participant_responses['participant'] = participant_responses['|__pMail'].str.split('@',expand=True)[0].str.lower()
participants['participant'] = participants['Participant ID'].str.split('@',expand=True)[0].str.lower()
```

```python
participants
```

```python
survey_participants = pd.merge(left=participant_responses, right=participants, left_on='participant', right_on='participant')
```

```python
survey_participants
```

```python
survey_responses_sad = pd.merge(left=survey_participants,right=song_features_sad,left_on='MUSIC SET',right_on='SET')
survey_responses_happy = pd.merge(left=survey_participants,right=song_features_happy,left_on='MUSIC SET',right_on='SET')
survey_responses_energetic = pd.merge(left=survey_participants,right=song_features_energetic,left_on='MUSIC SET',right_on='SET')

```

```python
survey_responses_sad.to_csv('survey_responses_chinese_sad.csv',encoding='utf-8')
survey_responses_happy.to_csv('survey_responses_chinese_happy.csv',encoding='utf-8')
survey_responses_energetic.to_csv('survey_responses_chinese_energetic.csv',encoding='utf-8')
```

```python
survey_responses_sad = pd.read_csv('survey_responses_english_sad.csv')
survey_responses_happy = pd.read_csv('survey_responses_english_happy.csv')
survey_responses_energetic = pd.read_csv('survey_responses_english_energetic.csv')
```

```python
distinct_sets = survey_responses_sad['SET'].unique()
len(distinct_sets)
```

```python
song_scores_sad = {}
song_scores_happy = {}
song_scores_angry = {}
song_scores_fear = {}

def cal_sad_score():
    for s in distinct_sets:
        df = pd.DataFrame()
        df = survey_responses_sad.loc[(survey_responses_sad['SET']==s)]
        df.reset_index(inplace=True)
        delta_happy = []
        delta_sad = []
        delta_angry = []
        delta_fear = []
        for i in range(len(df)):
            delta_happy.append(df['ans13a'][i]-df['ans12a'][i])
            delta_sad.append(df['ans13b'][i]-df['ans12b'][i])
            delta_angry.append(df['ans13c'][i]-df['ans12c'][i])
            delta_fear.append(df['ans13d'][i]-df['ans12d'][i])
        happy = sum(delta_happy)/len(delta_happy)
        sad = sum(delta_sad)/len(delta_sad)
        angry = sum(delta_angry)/len(delta_angry)
        fear = sum(delta_fear)/len(delta_fear)
        song_scores_sad[s]=sad
        song_scores_happy[s]=happy
        song_scores_angry[s]=angry
        song_scores_fear[s]=fear
```

```python
cal_sad_score()
song_features_sad['sad_score'] = song_features_sad['SET'].map(song_scores_sad)
song_features_sad['happy_score'] = song_features_sad['SET'].map(song_scores_happy)
song_features_sad['angry_score'] = song_features_sad['SET'].map(song_scores_angry)
song_features_sad['fear_score'] = song_features_sad['SET'].map(song_scores_fear)
```
